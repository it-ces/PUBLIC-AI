```python
## Iván Andrés Trujillo
## Variance unbiased estimator
import numpy as np
import matplotlib.pyplot as plt
mu, sigma = 40, 6.4
population = np.random.normal(mu, sigma, 5000)
np.var(population, ddof=0) # Population parameter
```

```python
sample_size = 10
trials = 9999
biased, unbiased = [], []
for _ in range(trials):
  sample = np.random.choice(population,sample_size)
  biased.append(np.var(sample, ddof=0)) # Divide by N
  unbiased.append(np.var(sample, ddof=1)) # Divide by N-1
plt.title(f"$\sigma^{2} = ${round(np.var(population, ddof=0),4)} ,  $S^{2}_{r'B'}$ = {round(np.mean(biased),4)} ,  $S^{2}_{'U'}$ = {round(np.mean(unbiased),4)} ")
plt.hist(biased)
plt.hist(unbiased);
```
