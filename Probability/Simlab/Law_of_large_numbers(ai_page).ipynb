```python
import random
import matplotlib.pyplot as plt
```

```python
# flip the coin
random.randint(0,1)
```

```python
proportions = []
for n in range (2, 1000):
 proportions.append(sum([random.randint(0,1) for _ in range(n)])/n)
```

```python
plt.plot(range(2,1000), proportions)
plt.hlines(0.5, xmin=2,  xmax=1000, color='red')
```

Therefore $\mu = 3.5$

```python
means = []
for n in range(2, 1000):
  means.append(sum([random.randint(1,6) for _ in range(n)])/n)
```

```python
plt.plot(range(2,1000), means)
plt.hlines(3.5, xmin=2,  xmax=1000, color='red')
```
