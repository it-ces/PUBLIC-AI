```python
import random
import matplotlib.pyplot as plt
```

```python
# flip the coin
random.randint(0,1)
```

```python
proportions = []
for n in range (2, 1000):
 proportions.append(sum([random.randint(0,1) for _ in range(n)])/n)
```

```python
plt.plot(range(2,1000), proportions)
plt.hlines(0.5, xmin=0,  xmax=1000, color='red')
```
