```python
import random
import matplotlib.pyplot as plt
import numpy as np
means = []
size =32
samples = 100000
for _ in range(samples):
  means.append(sum([random.randint(1,6) for _ in range(size)])/size)  # only works in this case
plt.hist(means)
```

```python
## Coming from binomial population
pop = [0,1]
trials = 5000
means = []
for _ in range(trials):
  means.append(np.random.choice(pop, size=60).mean())
plt.hist(means, bins=10)
```
