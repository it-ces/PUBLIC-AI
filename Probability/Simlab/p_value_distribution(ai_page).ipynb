```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import shapiro
```

```python
# Normal - mean - Pvalue
np.random.seed(321)
trials  = 2500
mu = 75.12
sd = 1
n_sample = 35
p_values = []
for i in range(trials):
  sample_mean = np.random.normal(mu, sd, n_sample).mean()
  p_values.append(1 - norm.cdf(x = sample_mean,
                               loc = mu,
                               scale = sd/(n_sample)**0.5))
plt.hist(p_values);
```

```python
# testing shapiro
trials = 2000
mu = 10
sd = 4.5
sample = 50
p_values = []
for _ in range(trials):
  p_values.append(shapiro(np.random.normal(mu, sd, sample)).pvalue)
plt.hist(p_values);
```

```python
# testing p - value in beta1 linear regression
beta0 = 7.4
beta1 = 0
sigma = 12
sample_size = 35
p_values = []
for _  in range(2500):
  x = np.random.normal(10, 8, sample_size)
  y = beta0 + beta1*x + np.random.normal(0, sigma, sample_size)
  b1_hat = ((x  - x.mean()) * (y - y.mean())).sum() / ((x  - x.mean())**2).sum()
  var_error = (sigma**2)/ (((x  - x.mean())**2).sum())
  p_values.append(2*(1 - norm.cdf(abs(b1_hat), loc = 0, scale =var_error**0.5))) # two tails beta1 different of zero
plt.hist(p_values);
```

```python
# Normal - mean - Pvalue
np.random.seed(321)
trials  = 2500
mu = 87.134
false_mu = 86.88
sd = 1
n_sample = 35
p_values = []
for i in range(trials):
  sample_mean = np.random.normal(mu, sd, n_sample).mean()
  p_values.append(1 - norm.cdf(x = sample_mean,
                               loc = false_mu,
                               scale = sd/(n_sample)**0.5))
plt.hist(p_values);
```
