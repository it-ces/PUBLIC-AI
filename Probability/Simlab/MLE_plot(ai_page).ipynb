```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

THETA = 0.27
pop = [1,0]
n_size = 600
data = np.random.choice(pop, p=[THETA, 1-THETA], size=n_size)
```

$$ L(\theta) = \binom{n}{k} \theta^{k} (1-\theta)^{n-k} $$

```python
def binomial(theta, success, trials):
  return math.comb(trials, success) * (theta**success) * (1-theta)**(trials-success)
```

```python
success = data.sum()
thetas = np.linspace(0,1,1000)
likelihoods  = [binomial(theta, success, n_size) for theta in thetas]
ThetaMLE = thetas[np.array(likelihoods).argmax()]
```

```python
plt.plot(thetas, likelihoods)
plt.title(r"$\hat{\theta}_{MLE}$ = " + f'{ThetaMLE.round(3)}');
```

Analytically we can $ d L(\theta) = 0$  and solve to get the maximun, but is easier $ d \log L(\theta) = 0$.

\begin{equation}
\log \binom{n}{k} + k \log (\theta) + (n-k) log(1-\theta)
\end{equation}

the derivative with respect to $\theta$ is:

\begin{equation}
\frac{k}{\theta} - \frac{n-k}{1 - \theta} = 0
\end{equation}

and therefore:

\begin{equation}
\hat{\theta}_{MLE} = \frac{k}{n}
\end{equation}
