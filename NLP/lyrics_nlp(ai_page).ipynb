```python
# Stopwords?
cancionario = ['adiós morenita me voy por la madrugada no quiero que me llores porque me da dolor',
               'la herida que siempre llevo en el alma no cicatriza',
               'noche de arreboles que incitan a ....',
               'un amor que se me fué otro amor que se me va por el mundo yo voy vagando',
               'ay ay ay ay la la la -- yo soy el aventurero el mundo me importa poco cuando una mujer me gusta',
               'quien fuera el búque de más potencia para  arrojarme al fondo del mar']
text = nlp(" ".join(cancionario))
[token.text for token in text if not token.is_stop and not token.is_punct]
```

```python
def clean_text(text):
  text = nlp(text)
  text = ' '.join([token.lemma_ for token in text if not token.is_stop and not token.is_punct])
  return text
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

df = pd.read_csv(url)

#vectorizer = TfidfVectorizer()
vectorizer = CountVectorizer()

X = vectorizer.fit_transform(df['lyrics'])

label_encoder = LabelEncoder()

y = label_encoder.fit_transform(df['label'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = MultinomialNB()

# Train the model
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Check unique labels in y and y_pred
print("Unique labels in y:", set(y))
print("Unique labels in y_pred:", set(y_pred))

# Get the class names from the label encoder
class_names = label_encoder.classes_

# Create a list of labels present in the predictions
# This is important to match target_names with the actual labels
labels_present = list(set(y_pred))

# Print class names and labels present for debugging
print("Class names:", class_names)
print("Labels present in predictions:", labels_present)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))

# Use labels parameter to match the labels in the classification report
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=class_names, labels=labels_present))
```

```python
# New song example
new_song = ["Cuando llegan las horas de la tarde Que me encuentro tan solo y muy lejos de ti... Me provoca volve' a los guayabalesDe aquellos sabanales donde te conocí..."]
# Transform the new song
new_song_transformed = vectorizer.transform(new_song)

# Predict the genre
predicted_genre = model.predict(new_song_transformed)

print(predicted_genre)
```

```python
# Preprocessing the data...

def pre_clean(string):
  string = string.lower()
  string = re.sub('  ', ' ', re.sub('\\n', ' ', string))
  string = re.sub('bis', ' ', re.sub('\\n', ' ', string))
  return string
```

```python
pre_clean(df.loc[0, 'lyrics'])
```

```python
clean_text(pre_clean(df.loc[0, 'lyrics']))
```

```python
df['lyrics'] = df['lyrics'].apply(lambda x : clean_text(pre_clean(x))) # still ay's and bis appear you must remove
```

```python
vectorizer = CountVectorizer()  # Using CountVectorizer instead of TF-IDF #token counts..

#vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(df['lyrics'])

label_encoder = LabelEncoder()

y = label_encoder.fit_transform(df['label'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = MultinomialNB()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Unique labels in y:", set(y))
print("Unique labels in y_pred:", set(y_pred))

class_names = label_encoder.classes_

labels_present = list(set(y_pred))

print("Class names:", class_names)
print("Labels present in predictions:", labels_present)

print("Accuracy:", accuracy_score(y_test, y_pred))

print("Classification Report:\n", classification_report(y_test, y_pred, target_names=class_names, labels=labels_present))
```
