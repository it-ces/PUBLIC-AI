```python
# loading the databases...
products = pd.read_csv("https://raw.githubusercontent.com/it-ces/clases-usco/main/products.csv")
orders = pd.read_csv("https://raw.githubusercontent.com/it-ces/clases-usco/main/order_products__train.csv")
```

```python
orders
```

```python
transactions = {}
for code in orders['order_id'].unique():
  transactions[code] = []
for index in orders.index:
  transactions[orders.loc[index, 'order_id']].append(orders.loc[index, 'product_id'])
```

```python
data = []
for t in transactions:
    data.append(transactions[t])
```

```python
with open('Twithproduct.csv','w') as f:
    for transaction in data:
        for item in transaction:
            f.write(str(item) + ',')
        f.write('\n')
```
