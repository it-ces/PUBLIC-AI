# Impact assesstment
This material is based in the book Bernal & Peña (2011).

In this section we are going to study, techniques to assess the contribution of a training or political program in a variable of interest.



##  ¿what would have happened?

What would have been the result in variable $y$ of the benefited individuals if they are not treated or trained?. The most important concept here is the conterfactual.


### Notation:
$D_{i}$ $i=0,1$ zero for individuals that not participated in the program and one in otherwise.

$y_{i}(j)$ $j=1,0$ the outcome varible for $i$  zero for individuals that not participated in the program and one in otherwise.

thus the effect over $i$ could be expressed as $\tau_{i} = y_{i}(1) - y_{i}(0)$.
Therefore, the average impact is the goal of study.


ATT (average treatment on the treated) is equal to $E[t_{i} | D_{i}=1]$.

\begin{equation}
E[y_{i}(1) \mid D_{i}=1] - E[y_{i}(0) \mid D_{i}=1]
\end{equation}

But both not are registered in database, we need estimate a good approximation of the counterfactual  $E[y_{i}(0) \mid D_{i}=1]$


# Random assignment is the ideal
The ideal is that the individuals are treated randomly. Therefore methodologies could be referred as  (quasi)experimental or nonexperimental.

Hint: if randomization is good then only we can see differences in the interest variable.
The following assumption (1*)
\begin{equation}
E[y_{i}(0) \mid D_{i} = 1 ]  -
E[y_{i}(0) \mid D_{i} = 0 ] = 0  
\end{equation}
allow us uses $E[y_{i}(0) \mid D_{i} = 0 ]$ as counterfactual, in theory not must be exist observed differences in individuals,  we can get this making a random assignment to the program.



if (1*) is true then we can estimate:


$\hat{\tau_{ATT}}  =  (\bar{y} \mid D=1) - (\bar{y} \mid D=0)$ the same result is gotten with  OLS

\begin{equation}
y_{i}  = \beta_{0} + \beta_{1} D_{i} + u_{i}
\end{equation}

If (1*) then $E(u_{i}\mid D_{i}) = 0$ then by OLS $\hat{\beta_{1}}$ is conistent and unbiased. in summary:
\begin{equation}
\hat{\beta_{1}} = (\bar{y} \mid D=1)  - (\bar{y} \mid D=0).
\end{equation}



## Bias by autoselection
(1*) no will by satisfy  if the participation in the program it by ownself selection. Therefore there is differecens in the observed a unobserved variables, even if the program does not exist, and therfore $E(u_{i} \mid D_{i}) \neq 0$ and $\hat{\beta_{1}}$ will be biased.




## Differences and differences (diff in diff)
An fortuite event could be produce a natural experiment, assesment of gliphosate effects of the health. but still could be difference in the base level of the chacteristics and in the interest variable among the groups.

$y_{t}|D=j$, where $t=1,2$ and $j=1,0$. $t=1$ before of treatment and $t=2$ after the treatment.

In natural experiments the assignment not could be perfect, and could there are difference among the groups that could lead to a bias the estimation of the impact.

Diff in diff is the difference among the groups of $(y_{2} \mid D=j) - (y_{1} \mid D=j)$ for $j=1,0$ namely;

\begin{equation}
\tau_{diff-diff} = (E[y_{2} \mid D=1] - E[y_{1} \mid D=1]) - (E[y_{2} \mid D=0] - E[y_{1} \mid D=0])
\end{equation}

In the sample $E[y_{t} \mid D=j] = (\bar{y_{t}} \mid D = j) $  where $\bar{y_{t}}$ is mean of the variable in $t$ period.


\begin{equation}
\tau_{diff-diff} = (\Delta \bar{y} \mid D=1) - (\Delta \bar{y} \mid D=0)
\end{equation}

if treatment is random then $\tau_{diff-diff}$ is unbiased.


Sometimes there are differences in the features of the individuls before
of treatment. Now with differences in differences we drop  initial levels of the variable among cases and controls.


\begin{equation}
\Delta y_{i} = \beta_{0}  + \beta_{1}D_{i} + v_{i}
\end{equation}

Now we can see that $E(u_{it} | D = j)= 0$ then; $E(\Delta y \mid D=1) = \beta_{0}  + \beta_{1}$ and $E[\Delta y | D = 0] = \beta_{0}$ and therefore:

\begin{equation}
E[\Delta  y \mid D = 1 ] - E[\Delta y \mid D = 0] = \beta_{1} = \tau_{diff-diff}
\end{equation}

```python
# Differences in Differences estimation.
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
df = pd.read_stata("experimentos_naturales_panel_base.dta")
df['diff'] = df['ha_nchs2'] - df['ha_nchs1'] # delta Y
```

```python
# difference estimation
print(smf.ols(formula = 'ha_nchs2 ~  + C(D)', data=df).fit().summary())

# Differences in Differences estimation
print(smf.ols(formula = 'diff ~  + C(D)', data=df).fit().summary())


# Differences in Differences adjusted
print(smf.ols(formula = 'diff ~  + C(D) + educa_jefe1 + ingresos_hogar_jefe1', data=df).fit().summary())
```
