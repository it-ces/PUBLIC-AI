```python
# defino la derivada
def dfquad(a,b,x):
  return 2*a*x + b
```

```python
dfquad(10,2,3)
```

```python
# Método Gradiente descendente
xi = 20
a = 2
b = 10
l_rate = 0.1 #hiperparámetro
for i in range (1000):
  xj = xi - l_rate*dfquad(a,b,xi)
  xi = xj 
print(xj)
```

```python
def fquad(a,b,c,x):
  return a*x**2 + b*x + c
```

```python
# debo sacar la segunda df de dfquad
def ddfquad(a):
  return 2*a
```

```python
# Método Raphson Newton
xi = 100
a = 2
b = 10
c = 0
for iterations in range(1000):
  xj = xi - (dfquad(a,b,xi)/ddfquad(a))
  xi = xj
  print(xi)
```
