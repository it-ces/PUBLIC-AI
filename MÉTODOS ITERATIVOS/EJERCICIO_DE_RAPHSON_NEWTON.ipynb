```python
# Definimos la función cuadrática
def fquad(a,b,c,x):
  return a*x**2 + b*x + c
```

```python
# Creamos una función que calcule su derivada
def dfquad(a,b,x):
  return 2*a*x + b
```

```python
# Generamos una función que encuentre el óptimo de las variables
def óptimo(a,b):
  return -b/(2*a)
```

1. Una pelota se lanza al aire desde una altura inicial de 2 metros y con una velocidad inicial de 10 m/s. La función que describe su altura en función del tiempo es h(t) = -4.9t^2 + 10t + 2. ¿Cuál es la altura máxima que alcanza la pelota y cuándo ocurre?

```python
tiempo_optimo=óptimo(-4.9,10)
tiempo_optimo # la pelota requiere un tiempo de 1.02seg para alcanzar la altura máxima
```

```python
fquad(-4.9,10,2,tiempo_optimo) # en 7.10 mtrs la pelota alcanza una altura máxima. 
```

```python
xi= 100
for iterations in range (1000):
  xj= xi -(fquad (-4.9,10,2,xi)/dfquad(-4.9,10,xi))
  xi=xj
  print(xi)
```

2. Una empresa fabrica y vende sombrillas. El costo total de producción de x sombrillas está dado por C(x) = 50x + 5000. El ingreso total por la venta de x sombrillas está dado por I(x) = 100x - 2x^2. Encuentra el número de sombrillas que la empresa debe producir y vender para maximizar su ganancia.

Se plantea la función de beneficios que biene dada por B= ingresos (I) - costos(C). de esta forma:

B= 100x-2x^2 -(50x+5000) B= 100x-2x^2-50x+5000 B= -2x^2 + 50x - 5000

De esta forma se obtiene la función de beneficios

```python
a,b,c= -2,50,-5000 
```

```python
beneficio=óptimo(-2,50)
beneficio # la empresa debe producir 12.5 sombrillas para maximizar su ganancia
```

```python
fquad(-2,50,-5000, beneficio) # -4687.5 son las sombrillas que se necesitan vender para maximizar la ganancia
```

```python
xi= 100
for iterations in range (1000):
  xj= xi -(fquad (-2,50,-5000,xi)/dfquad(-2,50,xi))
  xi=xj
  print(xi)
```

```python
xj # Aquí se alcanza la raíz de la función
```


3. Un proyectil se lanza desde el suelo con una velocidad inicial de 50 m/s y un ángulo de 30 grados respecto a la horizontal.
```python
a,b,c= -4.9,25,0
```

```python
distancia=óptimo(-4.9,25)
distancia # en 2.55 seg el proyectil alcanza su máximo vuelo
```

```python
fquad(-4.9,25,0, distancia) # en 31.88 mtrs el proyectil alcanza su máxima distancia
```

```python
xi= 100
for iterations in range (1000):
  xj= xi -(fquad (-4.9,25,0,xi)/dfquad(-4.9,25,xi))
  xi=xj
  print(xi)
```

```python
from matplotlib import pyplot
# Función cuadrática.
def f1(x):
    return -4.9*(x**2) + 10*x +2

# Valores del eje X que toma el gráfico.
x = range(-1000, 150000)
# Graficar ambas funciones.
pyplot.plot(x, [f1(i) for i in x])

# Establecer el color de los ejes.
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
# Limitar los valores de los ejes.
pyplot.xlim(-50, 50)
pyplot.ylim(-200, 100)
plt.plot(1.02,7.10,marker ="o")
plt.xlabel("Tiempo de pelota")
plt.ylabel("Altura")
# Guardar gráfico como imágen PNG.
pyplot.savefig("output.png")
# Mostrarlo.
pyplot.show()
```
