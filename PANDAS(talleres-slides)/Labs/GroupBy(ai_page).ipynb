```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'var1':[10,20,30,45],
                   'var2':[5,6,7,8]})
```

```python
df.agg(['sum', 'mean', 'min', 'max'])
```

```python
# We can build an aggregate funtion build by onself
```

```python
def cv(serie):
  array = serie.to_numpy()
  return array.std() / array.mean()
```

```python
df.agg([np.std, np.mean, cv])
```

```python
# We can index measures as Gini!
# 1 Excersice add gini as measure inequality!
```

```python
# Another way
```

```python
df.agg({'var1':[np.std],          # Only to a subset....
        'var2': [np.std, cv]})
```

```python
# agg with tuple..
df.agg(Mean = ('var1', np.mean),
       Sd = ('var2', np.mean))
```
