```python
URL = "https://raw.githubusercontent.com/it-ces/Datasets/main/MarketPower.csv"
```

```python
import pandas as pd
```

```python
df = pd.read_csv(URL)
```

```python
df['q'] = (df['theta'] - df['cm']) / 2*df['beta']
```

```python
df.head()
```

```python
df['sector'].value_counts()
```

```python
financial =  df[df['sector']=='Financial']
Q =  financial['q'].sum()
financial['share'] = financial['q'] / Q
```

```python
financial
```

```python
df.groupby(['sector'])['q'].sum()
```

```python
df['Total Q']  = df.groupby(['sector'])['q'].transform('sum')
```

```python
df['share'] = (df['q'] / df['Total Q'])*100
```

```python
df.groupby(['sector'])['share'].sum()
```

```python
def cv(serie):
  array = serie.to_numpy()
  return array.std() / array.mean()

# Gini.....
# Theil....
```

```python
df.groupby(['sector'])[['q', 'cm']].agg(['mean', 'std', cv])
```
