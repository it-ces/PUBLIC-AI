```python
import pandas as pd
```

```python
df = pd.DataFrame({'var1': [ 10, 20, 40, 33, 21],
      'var2': [1, 0, 1 ,0 ,1 ],
      'var3': ['A', 'B', 'C', 'D', 'E']})
df_ = df.copy()
```

```python
for var in df.columns:
  if df[var].dtypes.name == 'float64' or df[var].dtypes.name == 'int64':
    print(var)
    df[var] = (df[var] - df[var].mean())/ df[var].std()
```

```python
df # Var not requiere standardization
```

```python
def is_binary(var, df):
  ideal = [0,1]  # not binary.......
  flag = True
  for num in df[var].unique():
    if num not in ideal:
      flag = False
  return flag
```

```python
is_binary("var2", df=df_)
```

```python
is_binary("var2", df_)
```

```python
for var in df_.columns:
  if df_[var].dtypes.name == 'float64' or df_[var].dtypes.name == 'int64':
    if is_binary(var, df_) == False:
      df_[var] = (df_[var] - df_[var].mean())/ df_[var].std()
```
