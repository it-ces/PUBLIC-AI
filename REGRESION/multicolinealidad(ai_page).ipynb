```python
import numpy as np
def VIF(R):
  if R==1:
    return  np.inf
  else:
    return (1 / (1- R))
```

```python
url = "https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/profit-firms.csv"
```

# Financial accounting

 * equity translate to patrimonio,
 * asset activo.
 * asset =  liabilities + equity

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
```

```python
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.corr()
```

```python
# asset -> all goods that belong to the firm, cash, resources, properties , equipment, all "assests"
# equity -> asset - debt
```

```python
df['sopecho'] = df['equity'] + df['liabilities']
```

```python
df.head()
```

```python
import statsmodels.formula.api as smf
```

```python
print(smf.ols("profit ~ equity + asset + liabilities", data= df).fit().summary())
```

```python
print(smf.ols("profit ~ asset + liabilities", data= df).fit().summary())
```

```python
print(smf.ols("profit ~ asset + equity ", data= df).fit().summary())
```

```python
print(smf.ols("profit ~  equity  + liabilities", data= df).fit().summary())
```

```python
df.columns
```

```python
print(smf.ols('profit ~ advertising ', data = df).fit().summary())
```

```python
print(smf.ols('profit ~ expense ', data = df).fit().summary())
```

```python
print(smf.ols('profit ~ advertising + expense + equity', data = df).fit().summary())
```

```python
X = df[[  'expense',  'equity', 'asset' ]]
vif_data = pd.DataFrame()
X = sm.add_constant(X)
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
vif_data
```

```python
print(smf.ols(formula = 'profit ~ equity + expense + liabilities', data= df).fit().summary())
```

```python
print(smf.ols(formula = 'profit ~ equity + expense ', data= df).fit().summary())
```

```python
model = smf.ols(formula = 'equity ~  asset + expense ', data=df).fit()
```

```python
VIF(model.rsquared)
```
