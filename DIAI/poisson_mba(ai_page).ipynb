# Gestión de operación

Supongamos que en un hospital la cantidad de pacientes por hora sigue una distribución de poisson con parámetro es decir $\lambda$ pacientes por hora.

**cuantos médicos se necesitan, si cada médico atiende 4 pacientes por hora**


necesitamos:
$$ c \mu > \lambda $$
donde $c$ es el número de médicos, $\mu$ es la capacidad de atención del médico por hora, entonces:

$$ p = \frac{\lambda}{ c\mu }$$ si $p =1$ no hay margen para variabilidad, $p$ < 1 es mejor.
