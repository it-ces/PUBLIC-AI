El almacen de cadena los tres tristes tigres estima que la demana semanal de "gaseosa 100 ml" se distribuye
$$ X \sim N(600, 30)$$

el objetivo es que solamente que solamente el 5% de las semanas no se tenga stock, cuantas gaseosas debe pedir a la semana el gerente?



```python
from scipy.stats import norm
```

```python
norm.ppf(0.95, loc = 600, scale = 30)
```

```python
norm.cdf(649, loc = 600, scale = 30)
```
