por lo general lo importado es mas bárato pero eso puede gnerera costos de pérdida volatilidad, etc, nuestro objetivo es saber si es mejor o no en términos de costo.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy
```

```python
url = 'https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/supplier_cost_mba.csv'
```

```python
df = pd.read_csv(url)
```

```python
df
```

```python
df.groupby(by=['supplier'])['total_cost'].mean()
```

```python
scipy.stats.ttest_ind(df[df['supplier']=='domestic']['total_cost'],
                      df[df['supplier']=='imported']['total_cost'], equal_var=True ).pvalue
```
