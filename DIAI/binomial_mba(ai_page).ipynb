# marketing

Una empresa quiere realizar una campaña publicitaria
para eso con anterioriedad nos han suministrado la tasa de conversión (p), se presoupuesta llegar a (n)  personas
el costo por persona es de (c), la ganancia neta ( margen) por venta es de m, debemos calcular el ROI.

```python
# DATOS
p = 0.05 # probabilidad de conversión
n = 10000
c = 3000/n
m = 120
CT = c*n
```

lo que se modela son las ventas $X$ :

$$ X  \sim Binom(n,p) $$


El beneficio esperado es:

$$\pi_E =  E[X] m $$


el ROI es:

$$ \frac{ E[X]m  - CT}{CT} $$


```python
def ROI(n,p, m, c):
  ct = c*n
  return ((n*p*m) - ct )/ ct
```
