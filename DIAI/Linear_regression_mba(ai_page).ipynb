# For what is used linear regression?

in a linear model of the form
\begin{equation}
y = \beta_{0} + \beta_{1} x_{1} + u
\end{equation}

here denote the importance of $\beta_{1}$ we are interested in know **how increase $y$ with the increse of one unity in $x$**


```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
```

```python
url =  'https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/sales_advertising_reg1.csv'
```

```python
df = pd.read_csv(url)
```

```python
df.head()
```

```python
modelo = smf.ols('sales ~ advertising', data=df).fit()
print(modelo.summary())
```

# Important elements of output

**coef** is $\beta_{1}$, *p value* give us a confidence about the importance of the sign of the coefficient, given that is the probability not seen a zero in coef given the data, $R^{2}$ indicates how many our model explain ther variability of data.

# Multivariate model

We also can study how differente variables affect linearly a dependent variables


\begin{equation}
y = \beta_{0} + \beta_{1} x_{1} + \beta_{2} x_{2} + u
\end{equation}

```python
url2 = 'https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/sales_advertising_reg2.csv'
```

```python
df = pd.read_csv(url2)
```

```python
df.columns
```

```python
modelo = smf.ols('sales ~ advertising + popularity_index ', data=df).fit()
print(modelo.summary())
```
