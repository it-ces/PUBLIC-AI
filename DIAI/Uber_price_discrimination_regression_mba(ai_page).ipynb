## Modelo

$$
y = \beta_{0} + \beta_{1}D_{1} + \beta_{2}x + \beta_{3}(x D_{1}) + \epsilon
$$

$$
D_{1} =
\begin{cases}
1 & \text{si el individuo tiene la característica A} \\
0 & \text{en otro caso}
\end{cases}
$$

Donde $x$ es una variable continua.

---

## Valor esperado condicional

$$
E(y \mid x, D_1) =
\begin{cases}
\beta_{0} + \beta_{2}x & \text{si } D_1 = 0 \\
(\beta_{0} + \beta_{1}) + (\beta_{2} + \beta_{3})x & \text{si } D_1 = 1
\end{cases}
$$

---

## Derivadas (efecto marginal)

$$
\frac{\partial E(y \mid x, D_1)}{\partial x} =
\begin{cases}
\beta_{2} & \text{si } D_1 = 0 \\
\beta_{2} + \beta_{3} & \text{si } D_1 = 1
\end{cases}
$$

---

## Forma compacta

$$
\frac{\partial E(y \mid x, D_1)}{\partial x} = \beta_2 + \beta_3 D_1
$$

---

## Interpretación

- $\beta_0$: intercepto base  
- $\beta_1$: cambio en el intercepto cuando $D_1 = 1$  
- $\beta_2$: efecto de $x$ cuando $D_1 = 0$  
- $\beta_3$: cambio en la pendiente (interacción)

recordemos el concepto de elasticidad de demanda:

\begin{equation}
\epsilon =\frac{dQ}{dP}\frac{P}{Q}
\end{equation}

vamos a renombrar $\frac{dQ}{dP}$ as $b$

```python
def elasticity_linear_demand(a, b, price):
    """
    Q = a + bP  (b < 0)
    """
    Q = a + b * price
    return b * (price / Q)
```

```python
elasticity_linear_demand(41.06, -4.40, 5) # elastico
```

```python
elasticity_linear_demand(36.13, -0.22, 5) # inelastico!
```
