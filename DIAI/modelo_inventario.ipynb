# Recordemos un poco

Si yo tengo por ejemplo que determiar un inventario, tenemos que se generan dos costos:

el costo de ruptura y el costo de exceso, definimos por ejemplo que la demanda (D) y el inventario  o stock (I),
**el costo  de ruptura** se genera cuando:
  
\begin{equation}
D > I
\end{equation}

y el costo de exceso se genera cuando:

\begin{equation}
D < I
\end{equation}

si por ejemplo el costo unitario de adquisición es  de $c$, el precio es $p$  supongamos que el exceso lo podemos vender en sale en $s$ entonces el costo es:

\begin{equation}
C = max((D - I),0) ( p -c )  + max((I-D),0)(c-s)
\end{equation}

# la Demanda de audífonos X al mes se distribuye de la siguiente manera:

$$ X \sim N(\mu, sigma) $$


supongamos que $\mu = 200$, $\sigma = 20$.

Necesitamos el inventario, entonces supongamos que
el precio es de 4 dólares, y el costo unitario es de 2.5 dólares, el costo de ruptura es de:

```python
mu = 200
sigma = 20
p = 4
c = 2.5
s = 1.8 #precio de sale
```

```python
p - c # costo de ruptura o de oportunidad
```

```python
c - s # costo del exceso....
```

```python
def costo(p,c,s,D,I):
  # D es demanda
  # I es inventario!
  return (max(D-I,0) * (p - c))  + ( max(I-D,0) * (c-s))
```

```python
print(costo(p,c,s, D = 10, I = 9)) # que es más costoso ruputura?
print(costo(p,c,s,D = 9, I = 10))
```

```python
#  que pasa si yo compro satisfaciendo la mayor parte de la demanda
```

```python
I = norm.ppf(0.99, loc = mu, scale = sigma )
I
```

```python
norm.cdf(232, loc = mu, scale = sigma)
```

```python
trials = 100000
outcomes = []
for _ in range(trials):
  outcomes.append(costo(p,c,s,D = np.random.normal(loc = mu,
                                                   scale = sigma,
                                                   size = 1), I = I ))
sum(outcomes)/len(outcomes)
```

```python
# Economic Model
def I_opt(c,p, s, mu, sd):
 ratio = (p-c)/(p-s)
 return norm.ppf(ratio, loc = mu, scale = sd)
```

```python
def I_opt_ratio(c,p, s,):
 ratio = (p-c)/(p-s)
 return ratio
```

```python
1 - I_opt_ratio(c,p,s)
```

```python
norm.ppf(I_opt_ratio(c,p,s), loc = mu, scale = sigma)
```

```python
I_econ = I_opt(c,p,s,mu,sigma)
```

```python
trials = 100000
outcomes = []
for _ in range(trials):
  outcomes.append(costo(p,c,s,D = np.random.normal(loc = mu, scale = sigma, size = 1), I = I_econ ))
sum(outcomes)/len(outcomes)
```
