```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/it-ces/PUBLIC-AI/refs/heads/main/REGRESION/cobb-douglas-firm.csv"
```

```python
df = pd.read_csv(url)
```

```python
df
```

```python
import numpy as np
import statsmodels.api as sm
X = df[['K','L']]
X = sm.add_constant(X, prepend=False)
mod = sm.OLS(df['y'], X)
res = mod.fit()
print(res.summary())
```
