```python
import pandas as pd
import numpy as np
import itertools
import random
import math
from deap import base
from deap import creator
from deap import tools
from deap import algorithms

def profit(x):
  precios = [data[item][1] for item in data] # draw the prices
  acum = 0
  for i,j in zip(precios,x):
    acum = acum + i * j
  return acum

def feasiable(x, constrain):
  weights = [data[item][0] for item in data] # draw the prices
  acum = 0
  for i,j in zip(weights,x):
    acum = acum + i * j
  return acum<= constrain


def objective(x, constrain):
  if feasiable(x, constrain) == True:
    return profit(x),
  else:
    return -profit(x),
```

```python
# url soft
url_soft = "https://raw.githubusercontent.com/it-ces/Datasets/main/complex-soft.csv"
# url hard
url_hard = "https://raw.githubusercontent.com/it-ces/Datasets/main/complex-hard.csv"

df = pd.read_csv(url_hard)  # cnage complex soft or hard
display(df)
index =  ['a'+str(i) for i in range(df.shape[0])]
df.index  = index
hey =  df.to_dict('split')
data = {}
for article, info in zip(hey['index'], hey['data']):
  data[article] = info

CONSTRAIN = 23
```

```python
def select_method(method):
  if method == 'brute':
    all_sol = list(itertools.product([0, 1], repeat=len(data)))
    feasiable_items = {}
    for solution in all_sol:
      if feasiable(solution,CONSTRAIN):
        feasiable_items[solution] = profit(solution)
    optimun = sorted(feasiable_items.items(), key=lambda x:x[1], reverse=True)[0][0]

    print(list(np.array(list(data.keys()))[np.array(optimun, dtype='bool')]))
    print(profit(optimun)) # How many we earn?
    print(feasiable(optimun, CONSTRAIN))
  else:
    chromosomal_size=len(data)
    ### Define the problem
    creator.create("FitnessMax", base.Fitness, weights=(1.0,))
    creator.create("Individual", list, fitness=creator.FitnessMax)
    toolbox = base.Toolbox()
    toolbox.register("binary_gen", random.randint, 0, 1) # Gene values
    toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.binary_gen, chromosomal_size) # The length of chromosome!!
    toolbox.register("population", tools.initRepeat, list, toolbox.individual, 50)
    toolbox.register("evaluate", lambda x: objective(x,CONSTRAIN)) # Change constrain!
    toolbox.register("mate", tools.cxOnePoint)
    toolbox.register("mutate", tools.mutFlipBit, indpb=0.1) # The probablity of flip a gene!
    toolbox.register("select", tools.selTournament, tournsize=12)
    pop = toolbox.population()
    pop, logbok = algorithms.eaSimple(pop, toolbox, cxpb=0.5, mutpb=0.08, ngen=400, verbose=False)
    print(pop[0])
    print(np.array(list(data.keys()))[np.array(pop[0], dtype='bool')])
    print(profit(pop[0]))
    print(feasiable(pop[0], CONSTRAIN))
```
