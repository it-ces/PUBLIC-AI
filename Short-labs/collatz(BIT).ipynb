```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(14, 12))
def walk(n):
  if n==1:
    return 1
  elif n%2==0:
    return n / 2
  else:
    return 3*n + 1
def collatz(n):
  seq = []
  seq.append(n)
  while n!=1:
    n = walk(n)
    seq.append(n)
  return seq
for x in range(1,80):
  plt.plot(collatz(x))
```
