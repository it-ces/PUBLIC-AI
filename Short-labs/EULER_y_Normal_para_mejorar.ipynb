```python
#  https://www.programcreek.com/python/example/96643/matplotlib.animation.FuncAnimation

import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from IPython.display import HTML

# Taylor series of e**x
def e_taylor(loops, point):
    result = 0
    for i in range(loops):
        result  = result + ( point ** i / np.math.factorial(i))
    return result

def taylor_pol():
    # create a simple animation
    fig, ax = plt.subplots()
    line, = ax.plot([], [])
    plt.xlabel("iterations or terms added")
    plt.ylabel("$e^{x=4}$")
    ax.set_xlim(1, 30)
    ax.set_ylim(1, 10)

    def update(i):
      x = np.arange(1,i)
      y = [e_taylor(j,4) for j in x]
      line.set_data(x, y)
      plt.gca().set_title('Aproximation of $e^4= ${}'.format(e_taylor(i,4)))
      return line,

    return animation.FuncAnimation(fig, update, frames=30, interval=430, blit=True)
graph  = taylor_pol()
HTML(graph.to_jshtml())
```

```python
# Update in live a file
#importing libraries
import matplotlib.pyplot as plt
import matplotlib.animation as animation


fig = plt.figure()
#creating a subplot
ax1 = fig.add_subplot(1,1,1)

def animate(i):
    data = open('stock.txt','r').read()
    lines = data.split('\n')
    xs = []
    ys = []

    for line in lines:
        x, y = line.split(',') # Delimiter is comma
        xs.append(float(x))
        ys.append(float(y))


    ax1.clear()
    ax1.plot(xs, ys)

    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title('Live graph with matplotlib')


ani = animation.FuncAnimation(fig, animate, interval=1000)
plt.show()
```
