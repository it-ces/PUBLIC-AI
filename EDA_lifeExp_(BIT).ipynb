```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
URL = "https://raw.githubusercontent.com/it-ces/Datasets/main/wbdata-gdp-lifeExp.csv"
df = pd.read_csv(URL)
```

```python
# Histogram
plt.hist(df['lifeExp2020'], rwidth=1)
plt.savefig("Histogram.jpg")
```

```python
# Boxplot
df.dropna(inplace=True)
plt.boxplot(df['lifeExp2020'], vert=False)
plt.savefig("boxplot.jpg")
```

```python
sns.boxplot(df['lifeExp2020'])
plt.savefig("boxplot-sns.jpg")
```

```python
#  Comparing groups with boxplot
sns.boxplot(data=df, x="incomeLevel", y='lifeExp2020')
plt.savefig("boxplot-incomelevel.jpg")
```

```python
## TIME SERIE
URL = "https://raw.githubusercontent.com/it-ces/Datasets/main/wbdata-gdp-lifeExp-time-series.csv"
df = pd.read_csv(URL)
df  = df[df['value']=='Colombia'].reset_index(drop=True)
plt.plot(df['year'], df['lifeExp'])
plt.savefig("time-serie.jpg")
```
