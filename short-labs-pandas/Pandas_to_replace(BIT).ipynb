```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
    {'var1':[10, 22, 3, 44, np.nan, 4.1],
     'var2': ['Missing value', np.nan, 'Not value', 'Na', 'Missing', 'Unique'],
     'var3':['Black', 'White', 'Blue', 'Red', 'Yellow','Green']})
```

```python
# Create a New variable considering the color
# Put i a new variable one if is blue and zero in otherwhise
df['Blue'] = np.where(df['var3']=='Blue', 1, 0)
```

```python
# Create a variable using two conditions
df['White and Black'] = np.where((df['var3']=='White') | (df['var3']=='Black'), 1, 0 )
```

```python
# We are going to create a variable with numeric values
# create a variable with zero if the numer is lesser than 22 and 1 otherwise
df['Gter22'] = np.where(df['var1']<22, 0, 1)
```

```python
#### Consider that the before example have a mistake given that NaN is converted to one
## to consider this
df['Gter22_'] = np.where(df['var1'].isnull(), np.nan,
                         np.where(df['var1']<22, 0 ,1))
```

```python
## We want replace
# [Missing value, Not value, Na, missin] by NaN(np.nan)
df['var2'].replace(to_replace=["Missing value", "Not value", "Na", "Missing"],
                      value = np.nan)
```
