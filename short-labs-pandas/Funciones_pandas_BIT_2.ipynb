```python
# Funciones....
#Type hints
def cat(df,var,delimiter):
  """
  This function return a list  with unique values of a variable with a specific delimiter;
  ---------------------------------------------------------------------------------------
  Arguments
  ---------------------------------------------------------------------------------------
  df -> A dataframe class
  var -> is a string with the name of variable
  delimiter -> the character that separate the strings in var

  -------------------------------------------------------------------------------------------------
  return list
  """
  unique =[]
  for index in df.index:
    if type(df[var].iloc[index])==str:
      aux = df[var].iloc[index].split(delimiter)
      for cat in aux:
        if cat not in unique:
          unique.append(cat)
  return unique

def cats_gen (df,cats,var):
  """
  Return a dataframe with k-columns that contain 1 if the unique value  in cats is in var otherwise return 0.
  -----------------------------------------------------------------------------------------------------------
  Arguments
  ----------------------------------------------------------------------------------------------------------
  df -> A Dataframe class
  cats -> is a list
  var -> is the objective variable ( is the same var used to construct cats.)
 -------------------------------------------------------------------------------------------------------------
 return Dataframe
  """
  for column in cats:
    df[column]=0
    for index in df.index:
      if type(df[var].iloc[index])==str:
        if column in df[var].iloc[index]:
          df.loc[index,column]=1
  return df

import pandas as pd
def homogenizar(base):
  """
  Función que homogeniza los string de las variables
  """
  data = base.copy()
  for column in data.columns:
    if data[column].dtypes.name == 'object':
      for index in data.index:
        if type(df.loc[index,column]) == str:
          data.loc[index, column] = data.loc[index, column].upper().replace(" ","")
  return data

def dropLetters(base, var):
  """
  Función que elimina letras
  """
  data = base.copy()
  for row in data.index:
    if type(data.loc[row,var]) == str:
      data.loc[row,var] = re.sub("[íA-Za-z]+", "", data.loc[row,var])
  return data


def check(df):
  for col in df.columns:
    if df[col].dtypes.name=='object':
      print(col)
      print(df[col].value_counts())
      print('--'*100)
```

```python
def droplimitnan(df, treshold):
  N = df.shape[0]
  to_drop=[]
  for col in df.columns:
    nan = df[col].isna().sum()
    proportion_nan = nan/N
    if proportion_nan >= treshold:
      to_drop.append(col)
  df.drop(columns=to_drop, inplace=True)
  number = len(to_drop)
  return df
```
