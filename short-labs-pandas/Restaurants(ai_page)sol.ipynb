## Return on Assets (ROA)

Return on Assets (ROA) is a financial ratio that shows how profitable a company is relative to its total assets. It measures how efficiently a company is using its assets to generate net income.

\begin{equation}
ROA  = \frac{Net Income}{Total Assets}
\end{equation}

High ROA → The company is efficiently using its assets to generate profit.

Low ROA → The company might not be using its assets effectively.


```python
import pandas as pd
```

```python
URL = "https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/restaurants_usd.csv"
```

```python
df = pd.read_csv(URL)
```

```python
df
```

```python
df['net_income'] = df['total_income'] - df['total_cost']
df['ROA'] = df['net_income']/df['total_assets']
df.sort_values(by=['ROA'], ascending=False)[:5]
```

```python
df.drop(columns=['net_income',  'total_assets' ], inplace = True)
```
