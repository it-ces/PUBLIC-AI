```python
# Research what is sort_index?
def sum_dice(value, trials):
  outcomes = [np.random.randint(1,7, size=2).sum() for _ in range(trials)]
  acum = 0
  for result in outcomes:
    if result == value:
      acum +=1
  return acum/trials
```

# By firsr order expansion

A linearized way of $ln(x)$ by first order taylor expansion around = 1.

\begin{equation}
ln(x) \approx ln(1) + ln'(1)(x-1) \\
      \approx 0 + \frac{1}{1}(x-1)
\end{equation}

for $$u = \frac{(x + \Delta x)}{x} $$

Then we have that:

\begin{equation}
\ln \left(\frac{(x + \Delta x)}{x}\right) \approx \frac{\Delta x}{x} \\
\ln( x + \Delta x) - ln(x) \approx \frac{\Delta x}{x}
\end{equation}

Remeber that is a linear approximation therefore get have better results if you consider little changes.

```python
# Log daily return
df2['log_return']  = np.log(df2['Close']).shift(-1) - np.log(df2['Close'])
```

```python
df2['log_return'].hist() # Why appear symetrical?
```

```python
# Modelling with normal
# The real parameters are unkonwn...
mean = df2['log_return'].mean()
std = df2['log_return'].std()
```

```python
from scipy.stats import norm
domain = np.linspace(mean - 4*std, mean + 4*std, 100)
image = [norm.pdf(x, loc=mean, scale=std) for x in domain]
plt.plot(domain, image)
```

```python
# What is the probability of loss more than 3% in a day.
# The probability of log return is worse than 3%!
x0  = -0.03
```

```python
# Plot the area...
plt.plot(domain, image)
plt.ylim(norm.pdf(mean))
plt.fill_between(x=np.linspace(mean - 4*std, x0, 250),
                 y1=norm.pdf(np.linspace(mean - 4*std, x0, 250),
                 loc=mean, scale=std), y2=0)
```

```python
print("The area or probability is :", norm.cdf(-0.04, loc=mean, scale=std) )
```

# Value at  Risk (VAR)

ppf or percent point function.

The var could be useful to measure the money needed to tackle monetary losses.

```python
var = 0.05 # VAR to 95%
print(norm.ppf(var, loc = mean , scale = std))
# how much might lose given a probability value.
# How interpret it?
# 5% chance that daily return is worse than -0.020 (2 percent)
```
