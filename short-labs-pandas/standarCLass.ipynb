```python
# own functions
def z_score(var,df):
  return (df[var] - df[var].mean()) / df[var].std()

def is_binary(var, df):
  ideal = [0,1]  # not binary.......
  flag = True
  for num in df[var].unique():
    if num not in ideal:
      flag = False
  return flag

import pandas as pd
import numpy as np
url = "https://raw.githubusercontent.com/it-ces/labHealth/master/database.csv"

df = pd.read_csv(url)   # read_excel()
casa = [np.random.choice([0,1]) for _  in range(df.shape[0])]
childrens  = [np.random.choice([0,1,3,7,10,4]) for _  in range(df.shape[0])]
df['home'] = casa
df['childrens'] = childrens
```

```python
# Estandarizar Z Media 0 y su sd 1.......
```

```python
# El siguiente consistirá en aplicar la estadarización a las variables correspondientes..
# Que requiero???
# requiero --> aplicar la función z_score a las variables que debemos...
# le puedzo aplicar a una variable object?
```

```python
for column in df.columns:
  if (df[column].dtypes.name == 'float64' or df[column].dtypes.name=='int64') and is_binary(column,df)==False:
    df[column+'_z'] = (z_score(column, df)) #helpp!!!!!!!!!
    print(column)
```
