```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/product_sales_month_U.csv'
```

```python
df = pd.read_csv(url)
```

```python
df.isnull().sum()
```

```python
df[df.isnull().sum(axis=1).astype('bool')]
```

```python
df['share'] = df['sales']/df.groupby(by=['product'])['sales'].transform('sum')
df['sales_avg'] =  df.groupby(by=['product'])['sales'].transform('mean')
df['rank']= df.groupby(by=['product'])['sales'].rank(method='first')
rank = df[df['rank']<3].sort_values(by='product')
rank
```

```python
df[df['product']=='Granola bars']['sales'].mean()
```
