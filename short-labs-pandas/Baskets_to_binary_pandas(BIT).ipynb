```python
data = {'id':[1,2,3,4,5,6,7],
        'cestas':[
            ['leche','carne','frutas'], ['leche', 'tech'],
            ['frutas', 'ropa', 'arroz'] , ['leche', 'frutas'],
            ['leche', 'carne', 'pan'] , ['pan', 'tech'], ['leche']]
           }
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# sacar los únicos pósibles items...
def cats_binary(var,df):
  aux = []
  for cesta in df[var]:
    for product in cesta:
      if product not in aux:
        aux.append(product)
  for product in aux:
    df[product]= 0   # init all data in zero..
    for index in df.index:
      if product in df.loc[index,var]: # si encuentra ese producto en la cesta
        df.loc[index,product] = 1  # pongale 1 a esa columna en la transacción correspondiente.
  return df
```

```python
cats_binary('cestas',df)
```
