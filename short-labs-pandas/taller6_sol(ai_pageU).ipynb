```python
import pandas as pd
```

```python
url1="https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/sem1.csv"
url2="https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/sem2.csv"
```

```python
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
```

```python
full = pd.concat([df1,df2])
```

```python
full
```

```python
full['id'] = full['id-product'].apply(lambda x: x[0])
```

```python
full.groupby(by=['month','id']).count()
```
