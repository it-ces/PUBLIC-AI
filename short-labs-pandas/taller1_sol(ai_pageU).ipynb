```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/price-quantities-U.csv"
```

```python
df = pd.read_csv(url)
```

```python
df.columns
```

```python
df.drop(columns = ['ref_transaction', 'stock'], inplace=True)
```

```python
df.rename(columns={
    'price_registred_':'price',
    'month_extracted_from_Date':'month',
    'quantity_sold':'quantity'
}, inplace=True)
```

```python
df.sort_values(by=['quantity'], ascending=False)[0:5]
```

```python
df['promotion_day'].value_counts(normalize=True)
```

```python
df[df['month'] == 'December']['quantity'].mean()
```

```python
df['profit'] = df['quantity'] *  (df['price'] - df['unitary_cost'])
```

```python
df[df['month']=='January']['profit'].mean()  - df[df['month']=='December']['profit'].mean()
```
