```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/bankrupt-sector.csv"
```

```python
df = pd.read_csv(url)
```

```python
for sector in df['sector'].unique():
  print(sector,df[df['sector']==sector]['bankrupt'].value_counts())
```

```python
df['indicator']= 1
```

```python
df.groupby(by=['sector', 'bankrupt'])['indicator'].count()
```

```python
df.groupby('sector')['bankrupt'].value_counts(normalize=True)
```
