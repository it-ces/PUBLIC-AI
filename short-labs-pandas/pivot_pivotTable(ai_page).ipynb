```python
import numpy as np
df.pivot_table(index=['incomeLevel'], values=['NY.GDP.PCAP.PP.KD'], aggfunc=np.mean)
```

```python
df.pivot_table(index=['incomeLevel'], values=['NY.GDP.PCAP.PP.KD'], aggfunc="sum")
```

```python
df.pivot_table(index=['incomeLevel'], values=['NY.GDP.PCAP.PP.KD', 'NY.GNS.ICTR.ZS', 'SP.POP.GROW'], aggfunc="mean")
```

```python
df.pivot_table(index=['incomeLevel', 'time'], values=['NY.GDP.PCAP.PP.KD', 'NY.GNS.ICTR.ZS', 'SP.POP.GROW'], aggfunc="mean")
```

```python
def cv(serie):
  return serie.std()/serie.mean()
```

```python
# several funtions
df.pivot_table(index=['incomeLevel', 'time'], values=['NY.GDP.PCAP.PP.KD', 'NY.GNS.ICTR.ZS', 'SP.POP.GROW'], aggfunc=["mean", 'std', cv])
```

```python
data = {
    'Name': ['Shravan', 'Jeetu', 'Ram', 'Jeetu', 'Shravan', 'Ram'],
    'Subject': ['DSA', 'DSA', 'DSA', 'DBMS', 'DBMS', 'DBMS'],
    'Score': [85, 90, 75, 80, 88, 82],
    'Grade': ['A', 'A', 'B', 'B', 'A', 'B']
}
df = pd.DataFrame(data)
```
