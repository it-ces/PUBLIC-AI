```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/intro_dates.csv')
```

```python
df.head(2)
```

```python
df.info()
#observe that dates are strings (objects)!
```

```python
# we can transform in the object oriented to handle this type
df['Fecha_1'] = pd.to_datetime(df['Fecha_1'], format="%d/%m/%Y")
```

```python
df.info() # observe datetime64!
```

#Formats strings...

%d	Day of month 01-31	31

%B	Month name, full version	December

%m	Month as a number 01-12	12

%y	Year, short version, without century	18

%Y	Year, full version	2018

%j	Day number of year 001-366	365

%H hour, %M minute, %S second.


```python
df['Fecha_2'] = pd.to_datetime(df['Fecha_2'], format='%d/%m/%Y %H:%M:%S')
```

```python
# get information!
df['día2'] = df['Fecha_2'].dt.day
df['mes2'] = df['Fecha_2'].dt.month
df['mes_name2'] = df['Fecha_2'].dt.month_name()
df['hora-fecha2'] =df['Fecha_2'].dt.hour
```

```python
url2 ="https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/intro_dates2.csv"
df  = pd.read_csv(url2)
```

```python
df.columns
```

```python
df['start_date'] = pd.to_datetime(df['start_date'], format='%d-%m-%Y')
df['end_date'] = pd.to_datetime(df['end_date'], format='%Y-%m-%d')
```

```python
(df['end_date'] - df['start_date']).dt.days
```
