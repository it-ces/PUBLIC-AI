```python
import pandas as pd
from tableone import TableOne
import numpy as np
```

```python
def is_binary(df_, nums):
    df = df_.copy()
    variables = []
    for var in nums:
        flag = True
        unique = df_[var].unique()
        for value in unique:
            if value not in [0, 1, np.nan, 0.0, 1.0]:
                flag = False
        if flag == True:
            variables.append(var)
    return variables

def breakdown_vars(df, off_binary=False):
    """
    This function allow us categorize accodign to numerical or not
    """
    binaries = is_binary(df, df.columns)
    categorial = []
    nonormal = []
    normal = []
    for t in df.columns:
        if off_binary == False:
          if (df[t].dtypes.name=="object" or df[t].dtypes.name=='category') and  t not in binaries:
            categorial.append(t)
        else:
           if (df[t].dtypes.name=="object" or df[t].dtypes.name=='category'):
            categorial.append(t)
        if (df[t].dtypes=="int64" or df[t].dtypes=="float64") and t not in binaries:
                n,p = stats.shapiro(df[t])
                if p<0.05:
                    nonormal.append(t)
                else:
                    normal.append(t)
    if off_binary == False:
      return categorial, binaries, nonormal, normal
    else:
      return categorial, nonormal, normal
```

```python
url = "https://raw.githubusercontent.com/it-ces/Datasets/main/basesuper.csv"
df_ = pd.read_csv(url)
df = df_[['Ganancia (pérdida)', 'Pasivos corrientes totales',
    'Clasificación Industrial Internacional Uniforme Versión 4 A.C',
      'Activos corrientes totales','Patrimonio total','y']].copy()
```

```python
df['sector'] = df['Clasificación Industrial Internacional Uniforme Versión 4 A.C'].apply(lambda x: str(x)[0])
df['y'] = df['y'].astype('object')
```

```python
df.drop(columns=['Clasificación Industrial Internacional Uniforme Versión 4 A.C'], inplace=True)
```

```python
cat, nonormal ,  normal = breakdown_vars(df,  off_binary=True)
```
