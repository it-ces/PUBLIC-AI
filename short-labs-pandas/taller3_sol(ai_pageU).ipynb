```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/Spent_ref_customer-U.csv"
```

```python
df = pd.read_csv(url)
```

```python
df
```

```python
df['ref'] = df['ref_product'].apply(lambda x: x[0])
```

```python
df.groupby(by=['ref'])['total_spent'].sum().sort_values(ascending=False)[0:3]
```
