```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/it-ces/Datasets/main/basesuper.csv"
df = pd.read_csv(url)
```

```python
X = [ 'Activos corrientes totales',   'Total de activos',  'Ganancia (pérdida)'  , 'Total pasivos' ,  'Ingresos de actividades ordinarias' ]
```

```python
df = df[X]
df.corr() #You could change method='spearman'
```

```python
ax = sns.heatmap(df.corr(), annot=True, cmap='gray')
```

```python
corrMatrix = df.corr(method='spearman')
import numpy as np
mask = np.zeros_like(corrMatrix)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(12, 8))
    ax = sns.heatmap(corrMatrix, mask=mask, square=True, annot=True, cmap="gray")
```

```python
corrMatrix = df.corr(method='spearman')
import numpy as np
CorrMatrix = corrMatrix.where(np.tril(np.ones(corrMatrix.shape)).astype(bool))
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(12, 8))
    ax = sns.heatmap(CorrMatrix, annot=True, cmap="gray_r")
```
