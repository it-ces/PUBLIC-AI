```python
# huila
datos = {'year':[2001,2002,2003,2004,2005,2006], 'pib':[10,10.5, 11.7, 12.4,12.9,15]}
df = pd.DataFrame(datos)
```

```python
df # la tasa de crecimiento promedio de este señor..
df['crecimiento'] = 0
for index in df.index[0:-1]:
  df.loc[index, 'crecimiento'] = (df.loc[index,'pib']-df.loc[index+1,'pib'])/df.loc[index,'pib']
```
