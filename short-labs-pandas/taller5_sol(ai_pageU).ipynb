```python
import pandas as pd
```

```python
online = "https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/clothing-store-online.csv"
physical = "https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/clothing-store-physical.csv"
prices = "https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/clothing-store-prices.csv"
```

```python
online_df = pd.read_csv(online)
physical_df = pd.read_csv(physical)
prices_df = pd.read_csv(prices)
```

```python
# The total number of transaction could be in only one dataframe
online_df.head()
```

```python
physical_df.head()
```

```python
transactions = pd.concat([online_df, physical_df])
```

```python
transactions.head()
```

```python
prices_df.head()
```

```python
transactions.columns
```

```python
prices_df.columns
```

```python
df = transactions.merge(prices_df,
                   left_on='id-product',
                   right_on='reference', indicator=True)
```

```python
df
```

```python
df['expense'] = df['price'] * df['quantities']
```

```python
df['total_expense'] = df.groupby(by=['id-client'])['expense'].transform("sum")
```

```python
df
```

```python
df = df[-df.duplicated(['id-client'], keep='first')].reset_index(drop=True)
```
