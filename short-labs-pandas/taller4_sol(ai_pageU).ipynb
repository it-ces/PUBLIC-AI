```python
import pandas as pd
```

```python
url1 = "https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/days_overdue_m1.csv"
url2 = "https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/days_overdue_m2.csv"
url3 = "https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/days_overdue_m3.csv"
```

```python
month1 = pd.read_csv(url1)
month2 = pd.read_csv(url2)
month3 = pd.read_csv(url3)
```

```python
df  = month1.merge(month2, on='id_credit').merge(month3, on='id_credit')
```

```python
df.columns
```

```python
df['total_days'] = df[['days_overdue_month1', 'days_overdue_month2',
       'days_overdue_month3']].sum(axis = 1)
```
